# Question 2 ( Binominal Distribution) 

```python
#Question 2 
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
p=.05
n=10
# 2.A Probability that none of the items are defective?
r=0
a = stats.binom.pmf(r,n,p)
print("Probability that none of the items are defective",a)
#2.B Probability that exactly one of the items is defective?
r = 1
b= stats.binom.pmf(r,n,p)
print("Probability that exactly one of the items is defective",b)
#2.C Probability that two or fewer of the items are defective?
c=stats.binom.pmf(0,n,p) + stats.binom.pmf(1,n,p) + stats.binom.pmf(2,n,p)
print("Probability that two or fewer of the items are defective",c)
#2.D Probability that three or more of the items are defective ?
d = 1 - (stats.binom.pmf(0,n,p)+ stats.binom.pmf(1,n,p) + stats.binom.pmf(2,n,p)+ stats.binom.pmf(3,n,p))
print ("Probability that three or more of the items are defective",d)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
n=10
p=0.05
k=np.arange(0,8)
k
binomial = stats.binom.pmf(k,n,p)
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes')
plt.show()
```
